# Cleaning the data

Capitalize all the column names

```python
df.columns = df.columns.str.upper()
```

```python
df.head()
```

Generating GENDER column using SEX column, If value is 1 then MALE and If value is 2 then FEMALE

```python
df.loc[df['SEX']==1, 'SEX'] = 'MALE'
```

```python
df.loc[df['SEX']==2, 'SEX'] = 'FEMALE'
```

```python
df.loc[df['MARRIAGE']==1, 'MARRIAGE'] = 'MARRIED'
df.loc[df['MARRIAGE']==2, 'MARRIAGE'] = 'UNMARRIED'
df.loc[df['MARRIAGE']==3, 'MARRIAGE'] = 'OTHER'
```

```python
df.loc[df['EDUCATION'] == 1, 'EDUCATION'] = 'GRADUATE'
df.loc[df['EDUCATION'] == 2, 'EDUCATION'] = 'UNIVERSITY'
df.loc[df['EDUCATION'] == 3, 'EDUCATION'] = 'HIGH SCHOOL'
df.loc[df['EDUCATION'] == 4, 'EDUCATION'] = 'OTHERS'
```

```python
df['PAY_0'].replace([0,-1],'PAY DULY', inplace = True)
```

```python
df.head()
```

```python
df['PAY_2'].replace([0,-1,-2],['PAY DULY','PREPAID 1','PREPAID 2'], inplace = True)
```

```python
df['PAY_3'].replace([0,-1,-2],['PAY DULY','PREPAID 1','PREPAID 2'], inplace = True)
```

```python
df['PAY_4'].replace([0,-1,-2],['PAY DULY','PREPAID 1','PREPAID 2'], inplace = True)
```

```python
df['PAY_5'].replace([0,-1,-2],['PAY DULY','PREPAID 1','PREPAID 2'], inplace = True)
```

```python
df['PAY_6'].replace([0,-1,-2],['PAY DULY','PREPAID 1','PREPAID 2'], inplace = True)
```

```python
df.head()
```

```python
df = df.rename(columns={'PAY_0':'PAID_APRIL2005','PAY_2':'PAID_MAY2005','PAY_3':'PAID_JUNE2005','PAY_4':'PAID_JULY2005',
                        'PAY_5':'PAID_AUG2005','PAY_6':'PAID_SEP2005'})
```

```python
df.head()
```

```python
df = df.rename(columns={'DEFAULT PAYMENT NEXT MONTH':'DEFAULT_PAYMENT_NEXT_MONTH'})
```

```python
df.head()
```

```python
df = df.drop('DEFAULT_PAYMENT_NEXT_MONTH', axis = 1)
```

```python
df.head()
```

```python
df = df.rename(columns={'PAY_AMT1':'BILL_PAID_SEPT2005','PAY_AMT2':'BILL_PAID_AUG2005','PAY_AMT3':'BILL_PAID_JULY2005',
                        'PAY_AMT4':'BILL_PAID_JUNE2005','PAY_AMT5':'BILL_PAID_MAY2005','PAY_AMT6':'BILL_PAID_APRL2005'})
```

```python
df.head()
```

```python
df = df.rename(columns={'BILL_AMT6': 'BILL_APRL2005', 'BILL_AMT5': 'BILL_MAY2005','BILL_AMT4': 'BILL_JUNE2005', 'BILL_AMT3': 'BILL_JULY2005',
                        'BILL_AMT2': 'BILL_AUG2005', 'BILL_AMT1': 'BILL_SEP2005'})
```

```python
df.head()
```

```python
df['TOTAL_BILL'] = df[['BILL_SEP2005','BILL_AUG2005','BILL_JULY2005','BILL_JUNE2005',
                       'BILL_MAY2005','BILL_APRL2005']].sum(axis=1)
```

```python
df['TOTAL_BILL_PAID'] = df[['BILL_PAID_SEPT2005','BILL_PAID_AUG2005','BILL_PAID_JULY2005','BILL_PAID_JUNE2005',
                            'BILL_PAID_MAY2005','BILL_PAID_APRL2005']].sum(axis=1)
```

```python
df.head()
```

```python
df['DIFF_PERCENTAGE'] = (df['TOTAL_BILL_PAID'] / df['TOTAL_BILL']) * 100
```

```python
df.head(10)
```

```python
def get_grade(DIFF_PERCENTAGE):
    if DIFF_PERCENTAGE >= 75:
        return 'STANDARD'
    elif DIFF_PERCENTAGE >= 50 and DIFF_PERCENTAGE < 75:
        return 'SUBSTANDARD 1'
    elif DIFF_PERCENTAGE >= 25 and DIFF_PERCENTAGE < 50:
        return 'SUBSTANDARD 2'
    else:
        return 'DEFAULTER'

df['CLASSIFICATION'] = df['DIFF_PERCENTAGE'].apply(lambda x : get_grade(x))
```
